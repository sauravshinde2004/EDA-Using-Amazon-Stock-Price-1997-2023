# Moving Averages:

Plotting moving averages (e.g., 50-day, 100-day) of the Close price to identify trends.

```python
# Calculate moving averages
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA100'] = data['Close'].rolling(window=100).mean()

# Plotting the Close price and moving averages
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')
plt.plot(data['Date'], data['MA50'], label='50-Day Moving Average', color='orange')
plt.plot(data['Date'], data['MA100'], label='100-Day Moving Average', color='green')
plt.title('Amazon Stock Close Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

# Relative Strength Index (RSI) and Moving Average Convergence Divergence (MACD):

Plotting these technical indicators to identify potential buying or selling opportunities.

```python
# Calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculate MACD
def calculate_macd(data, short_window=12, long_window=26):
    short_ema = data['Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = data['Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal_line = macd.ewm(span=9, adjust=False).mean()
    return macd, signal_line

# Plotting RSI and MACD
plt.figure(figsize=(14, 10))

# RSI
plt.subplot(2, 1, 1)
plt.plot(df.index, calculate_rsi(df), label='RSI', color='blue')
plt.axhline(70, color='r', linestyle='--')
plt.axhline(30, color='g', linestyle='--')
plt.title('Relative Strength Index (RSI)')
plt.legend()

# MACD
plt.subplot(2, 1, 2)
macd, signal_line = calculate_macd(df)
plt.plot(df.index, macd, label='MACD', color='red')
plt.plot(df.index, signal_line, label='Signal Line', color='blue')
plt.title('Moving Average Convergence Divergence (MACD)')
plt.legend()

plt.tight_layout()
plt.show()
```

# Rolling Statistics:

Plotting rolling mean and standard deviation of the Close price to identify periods of volatility.

```python
# Calculate rolling mean and standard deviation
rolling_mean = data['Close'].rolling(window=30).mean()
rolling_std = data['Close'].rolling(window=30).std()

# Plotting rolling mean and standard deviation
plt.figure(figsize=(14, 8))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.plot(data.index, rolling_mean, label='Rolling Mean', color='red')
plt.fill_between(data.index, rolling_mean - rolling_std, rolling_mean + rolling_std, color='lightgray', label='Rolling Std Dev')
plt.title('Rolling Mean and Standard Deviation of Close Price')
plt.legend()
plt.show()
```
